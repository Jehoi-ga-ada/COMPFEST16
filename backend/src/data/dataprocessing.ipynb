# Saving important annotations

```python
# standardizing jobtitles
def lower_all(string):
    return string.lower()

# get unique_job_titles
unique_job_titles = it_only_df['jobTitle'].apply(lower_all).unique()
unique_job_titles_df = pd.DataFrame(unique_job_titles, columns=['jobTitle'])
```

We are going to reduce the job titles into several unique general job titles in the IT world

```python
from sentence_transformers import SentenceTransformer, util
# We'll use a pretrained model for context understanding and similarity

general_categories = [
    "Software Developer/Engineer", 
    "Systems Analyst", 
    "Network Administrator", 
    "Database Administrator (DBA)", 
    "IT Support Specialist", 
    "Cybersecurity Analyst", 
    "Data Scientist", 
    "Cloud Engineer", 
    "DevOps Engineer", 
    "Web Developer", 
    "IT Project Manager", 
    "Systems Architect", 
    "Business Analyst", 
    "IT Consultant", 
    "Quality Assurance (QA) Tester/Engineer", 
    "IT Manager", 
    "Data Engineer", 
    "Information Security Manager", 
    "Technical Support Engineer", 
    "Artificial Intelligence/Machine Learning (AI/ML) Engineer"
]

# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode all the general categories and job titles into embeddings
category_embeddings = model.encode(general_categories, convert_to_tensor=True)
job_title_embeddings = model.encode(unique_job_titles, convert_to_tensor=True)

# We will use cosine similarity to classify them into general categories
cos_sim = util.pytorch_cos_sim(job_title_embeddings, category_embeddings)

unique_job_titles_df['General Category'] = [general_categories[sim.argmax().item()] for sim in cos_sim]
unique_job_titles_df.sort_values(by=['General Category'], inplace=True)

unique_job_titles_df.to_csv('it_job_titles_mapping.csv', index=False)
```

# Creating the dataset for prediction

```python
# Group by 'postedAt' and 'jobTitle', then count occurrences
job_title_counts = it_only_df.groupby(['postedAt'])['General Category'].value_counts()
job_title_counts
```

```python
# Reset index to convert the groupby result into a DataFrame
job_title_counts_df = job_title_counts.reset_index(name='count')
job_title_counts_df.head()
```

```python
job_title_counts_df = job_title_counts_df.pivot_table(index='postedAt',
                                          columns='General Category',
                                          values='count')
```

```python
# fill missing values with 0 -> meaning no job posting
job_title_counts_df.fillna(0, inplace=True)
```

```python
# making sure the data is time series
job_title_counts_df.index = pd.to_datetime(job_title_counts_df.index)
full_date_range = pd.date_range(start=job_title_counts_df.index.min(), end=job_title_counts_df.index.max())
job_title_counts_df = job_title_counts_df.reindex(full_date_range)
```

```python
# checking missing values since adding a full date range
# might show some missing values
job_title_counts_df.isna().sum()
```

> We have 16 missing dates -> this will be filled with 0 since it means the missing dates doesn't have any it job postings

```python
job_title_counts_df.ffill(inplace=True)
```

```python
# total job titles
len(job_title_counts_df.columns.values)
```

```python
# exporting job title counts
job_title_counts_df.to_csv('job_title_counts.csv')
```

# Plotting 30 days mean of data

```python
import matplotlib.pyplot as plt
job_title_counts_df.rolling(window=30).mean().plot(kind='line')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
import plotly.graph_objects as go

rolling_mean = job_title_counts_df.rolling(window=30).mean().dropna()

# Plotly figure
fig = go.Figure()

# Add traces for each job title
for column in rolling_mean.columns:
    fig.add_trace(go.Scatter(
        x=job_title_counts_df.index,
        y=rolling_mean[column].values.flatten(),
        mode='lines',
        name=column,
        text=rolling_mean[column].apply(lambda x: f'{x:.2f}'),
        hoverinfo='text'
    ))

# Update layout
fig.update_layout(
    title='Cumulative Sum with Rolling Mean',
    xaxis_title='Job Title',
    yaxis_title='Cumulative Count',
    legend=dict(x=1, y=1),
    hovermode='x unified'
)

# Show plot
fig.show()
```
